```python
a = np.array([20,20,20])
b = np.array([3])

bound = 1
norm0 = np.linalg.norm(a)
norm1 = np.linalg.norm(b)
print(norm0, norm1, norm0-norm1)

a2 = a * (norm1+bound)/norm0

print(a2, np.linalg.norm(a2))
```
